```python
# Top 10 Allowed source IPs
import logging
logging.basicConfig(filename="extractor.log",
                    filemode='a',
                    format='%(asctime)s,%(msecs)d %(name)s %(levelname)s %(message)s',
                    datefmt='%H:%M:%S',
                    level=logging.DEBUG)

logging.info("Top 10 Allowed source IPs:")
with open("app.log") as f:
    dict={}
    for line in f:
      words_list = line.split()
      src=words_list[len(words_list)-5]
      action=words_list[len(words_list)-1]

      if(action=="Allow"):
        if dict.get(src) == None:
          dict[src]=1
        else:
          a=dict[src]
          dict[src]=a+1

    x=sorted(dict.items(), key=lambda x:x[1])
    for i in x[len(x)-10:len(x)]:
      print(i)
      logging.info(i)
```

```python
#  Top 10 Denied users and Top 5 destination IPs for each user
logging.info("Top 10 Denied users:")
with open("app.log") as f:
    dict={}
    for line in f:
      words_list = line.split()
      usr=words_list[len(words_list)-2]
      action=words_list[len(words_list)-1]
       
      if(action=="Deny"):
        if dict.get(usr) == None:
            dict[usr]=1
        else:
            a=dict[usr]
            dict[usr]=a+1
    x=sorted(dict.items(), key=lambda x:x[1])
    for i in x[len(x)-10:len(x)]:
      print(i)
      logging.info(i)
y= x[len(x)-10:len(x)]
logging.info("Top 5 destination IPs for each user:")
for i in range (0,len(y)):
  with open("app.log") as f:
   
    dict={}
    usr=y[i][0]
    print("\n")
    print(usr)
    logging.info(usr)
    for line in f:
        
      words_list = line.split()
      dest= words_list[len(words_list)-4]
         
      user=words_list[len(words_list)-2]
      if(usr==user):
        if(dict.get(dest) == None):
            dict[dest]=1
        else:
            a=dict[dest]
            dict[dest]=a+1

    x=sorted(dict.items(), key=lambda x:x[1])
    z=x[len(x)-5:len(x)]
    
    for i in z:
      print(i[0],end=" , ")
      logging.info(i[0])
```

```python
# Percentage of Bypassed TCP services and the Top 5 services
logging.info("Percentage of Bypassed TCP services:")
with open("app.log") as f:
    x=0
    for line in f:
      words_list = line.split()
      action=words_list[len(words_list)-1]
      protocol=words_list[len(words_list)-3]
         
      if(action=="Bypass"  and protocol=="TCP"):
            x=x+1
           
    
    print (str((x/1000000)*100)+"%")
    logging.info(str((x/1000000)*100)+"%")

logging.info("the Top 5 services:")
with open("app.log") as f:
    dict={("TCP","Allow"):0,
          ("TCP","Deny"):0,
          ("TCP","Bypass"):0,
          ("TCP","Log-only"):0,
          ("UDP","Allow"):0,
          ("UDP","Deny"):0,
          ("UDP","Bypass"):0,
          ("UDP","Log-only"):0}

    for line in f:
          words_list = line.split()
          action=words_list[len(words_list)-1]
          protocol=words_list[len(words_list)-3]
          a=dict[(protocol,action)]
          dict[(protocol,action)]=a+1
           
x=sorted(dict.items(), key=lambda x:x[1])
for i in x[len(x)-5:len(x)]:
      print(i)
      logging.info(i)

```
